```python
import torchvision.transforms as transforms
from torchvision.datasets import CocoDetection

# Define any transformations you want to perform on the images
transform = transforms.Compose([
    transforms.ToTensor()
])

# Define the base paths
base_img_folder = '/notebooks/dataset/loco_new/images'
annotations_folder = '/notebooks/dataset/loco_new/annotations'

# annotation files
annotation_files = {
    'all': 'loco-all-v1.json',
    'subset-1': 'loco-sub1-v1-val.json',
    'subset-2': 'loco-sub2-v1-train.json',
    'subset-3': 'loco-sub3-v1-train.json',
    'subset-4': 'loco-sub4-v1-val.json',
    'subset-5': 'loco-sub5-v1-train.json'
}

# Create datasets
datasets = {}

# Load the complete set
datasets['all'] = CocoDetection(
    root=base_img_folder,
    annFile=f'{annotations_folder}/{annotation_files["all"]}',
    transform=transform
)

# Define the subsets
subsets = ['subset-1', 'subset-2', 'subset-3', 'subset-4', 'subset-5']
# Load all subsets
for subset in subsets:
    datasets[subset] = CocoDetection(
        root=base_img_folder,
        annFile=f'{annotations_folder}/{annotation_files[subset]}',
        transform=transform
    )


# You now have a dictionary 'datasets' where each entry is a separate dataset
# Example: datasets['subset-1'], datasets['all']
```

```python
import torch
import torchvision
from torch.utils.data import DataLoader
# data loader
batch_size = 1
num_workers = 4

for subset in subsets:
    data_loader = torch.utils.data.DataLoader(datasets[subset],
                                              batch_size=batch_size,
                                              shuffle=False,
                                              num_workers=num_workers
                                             )


    import numpy as np
    import matplotlib.pyplot as plt

    num_samples = len(data_loader.dataset)
    print(f'The dataset {subset} contains {num_samples} samples.')

    required_keys = {'id', 'image_id', 'category_id', 'segmentation', 'area', 'bbox', 'iscrowd', 'isbbox', 'color', 'metadata'}
    # Counter for number of images displayed
    count = 0
    count_stop = 2
    for images, labels in data_loader:
        #print("  checking id: " + str(labels[0]["id"].item()), end='');
        # image shape is [batch_size, 3 (due to RGB), height, width]
        img = torchvision.transforms.ToPILImage()(images[0])
        plt.imshow(img)
        plt.show()
        print(labels[0].keys())
        print(labels)
        if not required_keys.issubset(labels[0].keys()):
            print("incorrect, id: " + str(labels[0]["id"].item()));

        # Increment the counter
        count += 1

        # Break the loop after displaying 3 images
        if count >= count_stop:
            break
    print(f"check done, everything seems correct {subset}")
```
